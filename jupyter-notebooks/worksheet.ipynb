```python
import pandas as pd
import pandas_datareader as r
from datetime import datetime,timedelta
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from functools import reduce
import yfinance


```

```python
ticks=["AMD","AAPL","TSLA"]
i=yfinance.download(tickers=ticks,period="5d",interval="15m")
# k=(i['Adj Close']-i['Adj Close'].shift(1))/i['Adj Close']
k=(i['Adj Close']-i['Adj Close'].shift(1))/i['Adj Close']
k["price"]=i['Adj Close']['AMD']
k.dropna()
# k['AMD'].round(3).to_csv("ss.csv",index=False)
k['AMD'].cumsum()
fig= go.Figure(go.Scatter(x=k.index ,y=k['AMD'],name="X"))
fig.add_trace(go.Scatter(x=k.index ,y=k['price'],name="X1"))
fig.update_xaxes(
    rangeslider_visible=True,)
```

```python
end_dt = pd.to_datetime(datetime.now())
goback = 30
st_dt = end_dt - timedelta(days=goback)
st_dt,end_dt
```

```python
tickers=['AMD','AAPL','TSLA']
df = r.DataReader("AMD","yahoo",st_dt)
df.columns

```

```python
df['x']=(2*df.Close - df.High - df.Low)/((df.High - df.Low)*df.Volume)
```

```python
f = make_subplots(specs=[[{"secondary_y": True}]])
x = go.Scatter(x=df.index ,y=df['x'],name="X")
x1= go.Scatter(x=df.index ,y=df['High'],name="X1")

f.add_trace(x,secondary_y=True)
f.add_trace(x1,secondary_y=False)

```

```python
df['pct_change']=df['Adj Close'].pct_change(1)
df
```
